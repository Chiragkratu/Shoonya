```python
from NorenRestApiPy.NorenApi import  NorenApi
from threading import Timer
import pandas as pd
import time
import concurrent.futures

api = None


class ShoonyaApiPy(NorenApi):
    def __init__(self):
        NorenApi.__init__(self, host='https://api.shoonya.com/NorenWClientTP/', websocket='wss://api.shoonya.com/NorenWSTP/')        
        global api
        api = self
```

```python
#imports 
import logging
import os 
import pyotp
 
#enable debug to see request and responses
logging.basicConfig(level=logging.DEBUG)

#start of our program
token= os.getenv('token')
otp = pyotp.TOTP(token).now()


user    = os.getenv('shoonya_user')
pwd     = os.getenv('shoonya_pwd')
factor2 = otp
vc      = os.getenv('VC')
app_key = os.getenv('APIKEY')
imei    = os.getenv('IMEI')

api = ShoonyaApiPy()
# authentication login
def authentication_login(user, pwd, factor2, vc, app_key, imei):
    ret = api.login(userid=user, password=pwd, twoFA=factor2, vendor_code=vc, api_secret=app_key, imei=imei)
    print(ret)
    
# authentication logout
def authentication_logout():
    ret = api.logout()
    print(ret)

# order placement 
def place_order(buy_or_sell, product_type,exchange, tradingsymbol, quantity,
                discloseqty, price_type, price, trigger_price, retention,
                amo,remark):
    
    ret = api.place_order(buy_or_sell, product_type,exchange, tradingsymbol, quantity,
                        discloseqty, price_type, price, trigger_price, retention,
                        amo, remark)
    print(ret)


# modify order
def modify_order(orderno, exchange, tradingsymbol, newquantity,newprice_type,
                newprice, newtrigger_price, amo):
    ret = api.modify_order(orderno, exchange, tradingsymbol, newquantity,newprice_type,
                newprice, newtrigger_price, amo)
    print(ret)

# cancel order
def cancel_order(orderno):
    ret = api.cancel_order(orderno)
    print(ret)

# exit order
def exit_order(orderno, prd=None):
    ret = api.exit_order(orderno,prd)
    print(ret)

# single order history 
def single_order_history(orderno):
    ret = api.single_order_history(orderno)
    print(ret)

    
```

```python
# Watchlists 
def get_watchlists():
    ret = api.get_watch_list()
    print(ret)

def get_watchlistnames():
    ret = api.get_watch_list_names()
    print(ret)

def add_scrip_to_watchlist(wlname,scrips):
    ret = api.add_watch_list_scrip(wlname,scrips)
    print(ret)

def delete_scrip_to_watchlist(wlname,scrips):
    ret = api.delete_watch_list_scrip(wlname,scrips)
    print(ret)

def search_scrips(exchange,searchtext):
    ret = api.searchscrip(exchange,searchtext)
    print(ret)
```
